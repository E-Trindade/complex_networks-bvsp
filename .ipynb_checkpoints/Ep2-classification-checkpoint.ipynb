```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)

clf = SVC(gamma='auto')
clf.fit(X_train, y_train)
# test_model('SVC', clf, X_test, y_test)
y_pred = clf.predict(X_test)
    
t = X_test.copy()
t['label'] = y_pred
t = t.merge(full_dataset[['close']], left_index=True, right_index=True, how='inner')
t = t.reset_index()
t = t.rename(columns={'index': 'day'})

plt.figure(figsize=(35, 25))


palette = sns.diverging_palette(240, 10, n=t.label.unique().size)

ax = plt.subplot()
sns.scatterplot(x='day', y='close', data=t,
                hue='label', palette=palette,
                s=25, markers=['o'], edgecolor=None, legend=None)

ax.set_xlim(t.day.min(), t.day.max())
```
